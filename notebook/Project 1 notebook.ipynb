# Remove Duplicates

```python
df = df.drop_duplicates()
```

# Convert Date Format

```python
df["Date"] = pd.to_datetime(df["Date"])
```

# Fill Missing Values

```python
df["CouponCode"] = df["CouponCode"].fillna("Notuse the Coupon")
```

# Save Cleaned Dataset

```python
df.to_excel("/content/drive/MyDrive/Data Set/Cleaned_Dataset.xlsx", index=False)
```
